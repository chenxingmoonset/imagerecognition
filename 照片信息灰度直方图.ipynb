```python
# 导入所需的库
import cv2
import os
import matplotlib.pyplot as plt
%matplotlib inline

# 指定数据集路径
dataset_path = 'C:/Users/beichen/dataset/meihuameat'
os.chdir(dataset_path)
print(os.listdir())

for file in os.listdir(): # 遍历每张图像
# 读取本地图片文件
    img = cv2.imread(file)

# 获取图片的宽度和高度
    w = img.shape[1]
    h = img.shape[0]

# 打印图片的尺寸信息
    print(f'图片的宽度是{w}像素，高度是{h}像素')
    # 将图片转换为灰度图
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 绘制灰度图的直方图
    plt.hist(gray.ravel(), bins=256)
    plt.xlabel('灰度值')
    plt.ylabel('频数')
    plt.title('灰度图直方图')
    plt.show()
os.chdir('../')
```
